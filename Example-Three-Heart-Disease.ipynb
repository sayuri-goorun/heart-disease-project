```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# from sklearn.linear_model import LogisticRegression
# from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("heart-disease.csv")
df.head()
```

```python
df.target.value_counts()
```

```python
sns.countplot(x="target", data=df, palette="Blues")
plt.show()
```

```python
countNoDisease = len(df[df.target == 0])
countHaveDisease = len(df[df.target == 1])
print("Percentage of Patients With No Heart Disease: {:.2f}%".format((countNoDisease / (len(df.target))*100)))
print("Percentage of Patients With Heart Disease: {:.2f}%".format((countHaveDisease / (len(df.target))*100)))
```

```python
sns.countplot(x="sex", data=df, palette="mako_r")
plt.xlabel("Sex ( 0 = Female , 1 = Male)")
plt.show()
```

```python
countFemale = len(df[df.sex == 0])
countMale = len(df[df.sex == 1])
print("Percentage of Female Patients: {:.2f}%".format((countFemale / (len(df.sex))*100)))
print("Percentage of Male Patients: {:.2f}%".format((countMale / (len(df.sex))*100)))
```

```python
df.groupby('target').mean()
```

```python
pd.crosstab(df.age, df.target).plot(kind="bar", figsize=(20,6))
plt.title("Heart Disease Frequency for Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
# plt.savefig('HeartDiseaseAndAges.png')
plt.show()
```

```python
pd.crosstab(df.sex, df.target).plot(kind="bar", figsize=(15,9), color=['#1CA53B','#AA1111'])
plt.title("Heart Disease Frequency for Sex")
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.xticks(rotation=0)
plt.legend(["No heart disease", "Heart disease"])
plt.ylabel("Frequency")
# plt.savefig('HeartDiseaseAndAges.png')
plt.show()
```

```python
plt.scatter(x=df.age[df.target==1], y=df.thalach[(df.target==1)], c="red")
plt.scatter(x=df.age[df.target==0], y=df.thalach[(df.target==0)], c="green")
plt.legend(["Disease", "No disease"])
plt.xlabel("Age")
plt.ylabel("Maximum Heart Rate")
plt.show()
```

```python
pd.crosstab(df.fbs, df.target).plot(kind="bar", figsize=(15,9), color=['blue','lightblue'])
plt.title("Heart Disease Frequency According to FBS")
plt.xlabel("FBS - (Fasting Blood Sugar > 120 mg/ dl) (0 = False, 1 = True)")
plt.xticks(rotation=0)
plt.legend(["No heart disease", "Heart disease"])
plt.ylabel("Frequency of Disease or Not")
plt.show()
```

```python
pd.crosstab(df.cp, df.target).plot(kind="bar", figsize=(15,9), color=['pink','grey'])
plt.title("Heart Disease Frequency According to Chest Pain Type")
plt.xlabel("Chest Pain Type")
plt.xticks(rotation=0)
plt.ylabel("Frequency of Disease or Not")
plt.show()
```
